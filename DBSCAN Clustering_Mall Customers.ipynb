# **DBSCAN Clustering**

**DBSCAN - Density-Based Spatial Clustering of Applications with Noise**


- It is easy to understand the math intuition
- It is an unsupervised ML technique
- Sometime it is better than K-means clustering and Hierarchial Clustering
- It has its owen advantages and disadvantages
- Using this alogorithm, we will try to make up clusters from the given dataset



### **Important Components in DBSCAN Clustering**

1. Epsilon
2. Minimum points - Minimum number of points within distance Epsilon
3. Core point
4. Border point
5. Noise point

# **Steps Involved:**

1. Importing libraries
2. Getting the dataset
3. Import DBSCAN alogorithm from sklearn.cluster library
4. Applying DBSCAN algorithm
5. Fitting the model
6. Labelling the data
7. 

```
### Importing libraries

import numpy as np
import pandas as pd
```

```
### Getting the dataset

dataset = pd.read_csv('/content/drive/My Drive/MachineLearning/KNN/Mall_Customers.csv')
dataset.head(10)
```

```
x = dataset.iloc[:,[3,4]].values
```

```
x
```

```
### Applying DBSCAN alogorithm

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3, min_samples=4)
```

```
### Fitting the model

model = dbscan.fit(x)
```

```
### Plotting the clusters
import matplotlib.pyplot as plt

predict = model.fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=predict, cmap='Paired')
plt.title("DBSCAN")
```

```
### Labelling the data
labels = model.labels_

labels
```

```
### Displaying the unique values

labels_unique = set(labels)
labels_unique

### Outliers are -1 which have not fallen in one of the cluseters from Cluster 0 to Cluster 8
```

```
len(labels_unique) # Total number clusters and it includes -1 (outliers) which should not be included in the number of clusters, so the number of clusters is 9
```

```
### Identifying the points which make up our core points

sample_cores = np.zeros_like(labels, dtype=bool)
sample_cores
```

```
sample_cores[dbscan.core_sample_indices_] = True
```

```
sample_cores
```

```
### Calculating the number of clusters

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
```

```
n_clusters
```

## **Advantages**

1. It is great at separating clusters of high desity versus clusters of low density within a given dataset
2. It is great to able to find points not fitting into any clusters - It detects outliers


## **Disadvantages**

1. It does not work well when dealing with clusters of varying densities. DBSCAN is great at separating high density clusters from low density clusters, DBSCAN struggles with clusters of similar density
2. It sturggles with high dimensionality data
